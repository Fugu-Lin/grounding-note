```python
import google.generativeai as genai
```

```python
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# Select the model
model = genai.GenerativeModel(
    'gemini-1.5-pro-latest',
    generation_config=genai.GenerationConfig(
        max_output_tokens=2000,
        temperature=0.9,
    ))
```

## Use different prompt instructions to direct the answer output 
### Example 1: simple instruction

```python
chat = model.start_chat (history=[])

instruction = "Greet the user when initializing the conversation. Return the summary from googling according to users' query. "

while(True):
  question = input("User Input: ")
  if(question.strip() == ''):
    break
  response = chat.send_message(instruction + question)
  print('\n')
  print (f"Bot: {response.text}")
  print('\n')
  instruction = ''
```

## Use different prompt instructions to direct the answer output 
### Example 2: simple instruction

```python
chat = model.start_chat (history=[])

instruction = "Greet the user when initializing the conversation. Say goodbye when the user ends the conversation. Return the search summary from Google, YouTube, Medium, BBC News, PubMed, according to users' query. Return source URL and the image as well. Let the user know they can exit by typing STOP."

print("This is an AI tool search bot that can help you find information from Google, YouTube, Medium, BBC News, and PubMed. Please enter the keywords for your search in the box.")

while(True):
  question = input("Enter your questions here: ")
  if(question.strip() == 'STOP'):
    print("Goodbye!")
    break
  response = chat.send_message(instruction + question)
  print('\n')
  print (f"Bot: {response.text}")
  print('\n')
  instruction = ''
```
